```
import matplotlib.pyplot as plt
```

```
x=[3,8] #these are x1 and x2
y=[7,14] #these are y1 nd y2
y1=[8,15]
x1=[2,7]
```

```
plt.xlabel("time")
plt.ylabel("distance")
plt.plot(x,y,label="speed of car") #this will draw a line 
plt.plot(x1,y1,label="speed of truck")
plt.grid(color='green') #to draw grids
plt.legend()
plt.show()
#this is for line form representation
```

```
plt.xlabel("time")
plt.ylabel("distance")
plt.bar(x,y,label="speed of car") #this will draw a bar graph
plt.bar(x1,y1,label="speed of truck")
plt.plot(x,y,label="speed of car") #this will draw a line 
plt.plot(x1,y1,label="speed of truck")
plt.grid(color='green') #to draw grids
plt.legend()
plt.show()
```
