```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
players=["virat","dhoni","rohit","rahul"]
runs=[1200,200,800,700]
runs1=[1300,900,1400,3000]
```

```
fig,ax=plt.subplots(2)
ax[0].pie(runs,labels=players,shadow=True,autopct='%1.1f%%',explode=(0.2,0,0,0.1))
ax[0].legend()
ax[1].pie(runs1,labels=players,shadow=True,autopct='%1.1f%%',explode=(0.2,0,0,0.1))
ax[1].legend()

plt.show()
```
